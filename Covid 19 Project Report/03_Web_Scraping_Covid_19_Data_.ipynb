```
### Beautiful Soup - for 1 page
### Scraping - for whole website

from bs4 import BeautifulSoup as bs
import requests
```

```
page = requests.get('https://www.mohfw.gov.in')
page
```

```
page.content
```

```
soup = bs(page.content,'html.parser')
soup
```

```
# Extracting the table data
data = soup.find_all('div',class_='data-table table-responsive')
data
```

```
# Extracting all the tr tags from the tables
all_data = soup.find_all('tr')[1:]  # skipping the zeroth index and printing from the first one
all_data 
```

```
# Extracting all the td tags from the td
all_data_td = soup.find_all('td')[1:]  # skipping the zeroth index and printing from the first one
all_data_td 
```

```
# Extraciting useful information from the td tags
all_data_list = []
for i in range(0,len(all_data_td)):
  all_data_list.append(all_data_td[i].get_text())
all_data_list
```

```
states = []
for i in range(0,len(all_data_list),5):  
  states.append(all_data_td[i].get_text())
states = states[0:-3] # junk Information
states
```

```
confirmed = []
for i in range(1,len(all_data_list),5):  
  confirmed.append(all_data_td[i].get_text())
confirmed = confirmed[0:-3] # junk Information

cured = []
for i in range(2,len(all_data_list),5):  
  cured.append(all_data_td[i].get_text())
cured = cured[0:-3] # junk Information

deaths = []
for i in range(3,len(all_data_list),5):  
  deaths.append(all_data_td[i].get_text())
deaths = deaths[0:-2] # junk Information

print(len(states))
print(len(confirmed))
print(len(cured))
print(len(deaths))
```

```
import pandas as pd
df = pd.DataFrame()
df['States'] = states
df['Confirmed'] = confirmed
df['Cured'] = cured
df['Deaths'] = deaths
print(df)
```

```
df.to_csv('my_covid_india.csv')
```
